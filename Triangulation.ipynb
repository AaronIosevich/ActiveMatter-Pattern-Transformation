```python
from skfem import *
import numpy as np
from skfem import MeshTri
import matplotlib.pyplot as plt
from scipy.spatial import Delaunay
from scipy.spatial import distance

def generate_points(num_points, minimum_distance):
    points = []
    while len(points) < num_points:
        point = np.random.rand(2)
        if all(distance.euclidean(point, i) >= minimum_distance for i in points):
            points.append(point)
    return np.array(points)

num_points = 1200
minimum_distance = 0.0225
points = generate_points(num_points, minimum_distance)


tri = Delaunay(points)


mesh = MeshTri(points.T, tri.simplices.T)


plt.triplot(points[:, 0], points[:, 1], tri.simplices, color='blue')
plt.scatter(points[:, 0], points[:, 1], color='red',s=1.75)
plt.gca().set_aspect(1.0)
fig=plt.gcf()
fig.set_size_inches(20,20)
plt.title('Delaunay Triangulation')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
